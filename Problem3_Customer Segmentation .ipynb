```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {
   "Age": [18,22,25,30,35,40,45,50,55,60],
   "SpendingScore": [80,75,78,60,55,50,45,40,35,30]
}
df = pd.DataFrame(data)
df
```

```python
X = df[["Age", "SpendingScore"]]
model = KMeans(n_clusters=2, random_state=42)
model.fit(X)
df["Cluster"] = model.labels_
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Age"], df["SpendingScore"], c=df["Cluster"])

plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")
plt.show()
```

●	How many groups formed?
    2

●	Describe each group in business terms
    young high-spending customers and older low-spending customers, enabling targeted marketing strategies
