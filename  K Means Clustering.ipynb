```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {
   "Income": [20,22,25,30,35,40,42,45,48,50],
   "Spending": [30,35,28,40,45,50,52,55,60,65]
}
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df["Income"], df["Spending"])
plt.xlabel("Income")
plt.ylabel("Spending")
plt.show()
```

```python
model = KMeans(n_clusters=3)
model.fit(df)
```

```python
labels = model.labels_
print(labels)
```

```python
df["Cluster"] = labels
df
```

```python
plt.scatter(df["Income"], df["Spending"], c=df["Cluster"])
plt.xlabel("Income")
plt.ylabel("Spending")
plt.show()
```

If we increase the number of clusters to 3, the algorithm segments customers into low, medium, and high value groups, allowing for more granular marketing strategies
