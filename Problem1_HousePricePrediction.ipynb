```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {
   "Size_sqft": [500, 700, 900, 1100, 1300, 1500, 1700, 1900],
   "Price": [150000, 180000, 220000, 260000, 300000, 340000, 380000, 420000]
}
df = pd.DataFrame(data)
df
```

```python
X = df[["Size_sqft"]]
y = df["Price"]
```

```python
model = linear_model.LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[2100]])
print(prediction)
```

●	What price is predicted?
    457500

  
●	Does price increase linearly with size?
    YES it increased linearly


```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.title("Size_sqft vs Price")
plt.xlabel("Size_sqft")
plt.ylabel("Price")
plt.show()
```
